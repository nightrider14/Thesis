```
#Function to remove punctuations
def remove_punctuation(text):
    '''a function for removing punctuation'''
    import string
    # replacing the punctuations with no space, 
    # which in effect deletes the punctuation marks 
    translator = str.maketrans('', '', string.punctuation)
    # return the text stripped of punctuation marks
    return text.translate(translator)
#Function to stem
stemmer = SnowballStemmer("english")
def stemming(text):    
    '''a function which stems each word in the given text'''
    text = [stemmer.stem(word) for word in text.split()]
    return " ".join(text) 
#Function to remove stop words
sw=stopwords.words('english')
#Display stop words
np.array(sw)
#Function to remove sop words
def remove_StopWords(text):
    '''a function for removing the stopword'''
    # removing the stop words and lowercasing the selected words
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    # joining the list of words with space separator
    return " ".join(text)
```

```
embeddings_index = {}
f = open('glove.6B.300d.txt', encoding='utf-8')
embedding_values = {}
for line in tqdm(f):
    value = line.split(' ')
    word = value[0]
    coef = np.array(value[1:],dtype = 'float32')
    embedding_values[word] = coef
```

```
all_embs = np.stack(embedding_values.values())
emb_mean,emb_std = all_embs.mean(), all_embs.std()
emb_mean,emb_std
```

```
x = df['TEXT']
x2=df2['TEXT']
x3=df3['TEXT']
```

```
token = Tokenizer(num_words=24000)
token.fit_on_texts(x)
seq = token.texts_to_sequences(x)

token2 = Tokenizer(num_words=24000)
token2.fit_on_texts(x2)
seq2 = token2.texts_to_sequences(x2)

token3 = Tokenizer(num_words=24000)
token3.fit_on_texts(x3)
seq3 = token.texts_to_sequences(x3)
```

```
pad_seq = pad_sequences(seq,maxlen=500)
pad_seq2 = pad_sequences(seq2,maxlen=500)
pad_seq3 = pad_sequences(seq3,maxlen=500)
```

```
#length of vocablary
vocab_size = len(token.word_index)+1
print(vocab_size)
```

```
embedding_matrix = np.random.normal(emb_mean, emb_std, (vocab_size, 300))
for word,i in tqdm(token.word_index.items()):
    values = embedding_values.get(word)
    if values is not None:
        embedding_matrix[i] = values
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(pad_seq, y, test_size=0.20, random_state=42)
x_test2=pad_seq2
y_test2=y2
x_test3=pad_seq3
y_test3=y3
```

```
model1 = Sequential()
model1.add(Embedding(vocab_size,300,input_length=500,weights = [embedding_matrix],trainable = False))
model1.add(Bidirectional(LSTM(300,return_sequences=True)))
model1.add(GlobalMaxPool1D())
model1.add(Dense(300,activation = 'relu'))
model1.add(Dropout(0.2))
model1.add(Dense(5,activation = 'sigmoid'))
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model1.summary()
```

```
from keras.callbacks import EarlyStopping	
# simple early stopping
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1)
```

```
history = model1.fit(X_train,y_train,epochs =50,batch_size=16,validation_split=0.1,callbacks=[es])
```

```
def get_model():
    inp = Input(shape=(500, ))
    x = Embedding(vocab_size,300,input_length=500,weights = [embedding_matrix],trainable = False)(inp)
    x = SpatialDropout1D(0.2)(x)
    x = Bidirectional(GRU(80, return_sequences=True))(x)
    avg_pool = GlobalAveragePooling1D()(x)
    max_pool = GlobalMaxPooling1D()(x)
    conc = concatenate([avg_pool, max_pool])
    outp = Dense(5, activation="sigmoid")(conc)
    
    model = Model(inputs=inp, outputs=outp)
    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])

    return model
```
